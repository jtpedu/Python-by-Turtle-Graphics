実はPythonでは、while文による繰り返し処理ができます！

while文は、指定した条件式が真の間、処理を繰り返し実行します。

今度は🐢を頂点の間に5回移動させるため、変数iに1を初期値として代入し、

while文の条件としてiが5以下を指定します。

そして、ブロック内で変数iに代入されている値に1を加算します。

今回の処理の流れは👇のようになります。

指示に従って実行してみてください〜

```python
##########################################
# まず
# home()
# clear()
# を入力してリセットしましょう
##########################################

```

```python
##########################################
#👇に
# i = 1
# while i<=5:
#  right(144)
#  forward(100)
#  i = i+1
#と入力して実行してみてください。
#########################################





```

あれ❓❓❓★はちょっと歪んでいませんか！😺😺😺😺😺

大丈夫です！コードを少し編集すれば★を水平に保つことができますよ〜

👇の指示にしたがってみてください！

```python
##########################################
# まず
# home()
# clear()
# right(90)
# forward(100)
# を入力して1本目の線に戻りましょう。
##########################################




```

🐢さんは既に二つ目の頂点にたどり着いたので、

さらに5-1=4回の繰り返し処理をすれば五つ目の頂点に到達できますね。

```python
##########################################
#👇に
# i = 1
# while i<5:
#  right(144)
#  forward(100)
#  i = i+1
#と入力して実行してみてください。
#########################################




```
